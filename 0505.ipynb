```python
import numpy as np
plt.plot([1,2,3,4],[0,0.3,0.6,0.9], 'gx') #綠色
plt.plot([1,2,3,4],[0,0.3,0.6,0.9], 'r--') #紅線

#取亂數
X = 1+np.arange(30)/10
delta = +np.random.uniform(low=-0.1,high=0.1, size=(30,))
#帶公式
Y=0.3*X-0.3 + delta
plt.plot(X,Y,'bo') #藍點
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import pandas as pd
from sklearn import datasets,linear_model

dataframe = pd.read_fwf('brain_body.txt')  #讀取檔案
x_values = dataframe[['Body']] #身體數據
y_values = dataframe[['Brain']] #大腦數據

#訓練
body_reg = linear_model.LinearRegression() #使用線性回歸模型

body_reg.fit(x_values, y_values)
pre = body_reg.predict(x_values)

#訓練預測
print(body_reg.predict(pd.DataFrame(data=[[170]])))

plt.scatter(x_values, y_values)
plt.plot(x_values,pre)
plt.show()
```

```python
#取得糖尿病的數據
diabetes = datasets.load_diabetes()

#只取三個特徵值 BMI
diabetes_X = diabetes.data[:, np.newaxis,2]

#切分特徵值
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

#切分答案
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

#計算
regr = linear_model.LinearRegression()

#均方誤差
regr.fit(diabetes_X_train, diabetes_y_train)

#顯示方差分數 1表示完美預測
print('Coefficients: \n', regr.coef_)

print('Mean squared error: %.2f' %np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2))

print('Variance score: %.2f' %regr.score(diabetes_X_test, diabetes_y_test))

plt.scatter(diabetes_X_test, diabetes_y_test, color='black') #匯出黑點
plt.plot(diabetes_X_test, regr.predict(diabetes_X_test), color='blue', linewidth=3) #匯出測試預測的線性回歸答案

plt.xticks(())
plt.yticks(())

plt.show()
```
